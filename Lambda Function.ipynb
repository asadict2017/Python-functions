```python
#Lambda function: small anonymous function defined using the lambda keyword
#They can have any numbers of argurments but only one expression (means one logic)
#commonly used for short operations

#syntax: lambda arguments:expression/logic
```

```python
#addition for two numbers
```

```python
#Traditional function
def addition(a,b):
  return a+b
addition(2,3)
```

```python
#Lambda function
AddFunc=lambda a,b:a+b
type(AddFunc)
##calling
AddFunc(2,3)
```

```python
#another example for checking even number?
#Traditional way:
def evencheck(num):
  if num%2==0:
    return True
  else:
    return False
evencheck(12)
```

```python
#lambda ways
evencheck=lambda num:num%2==0
evencheck(11)
```

```python
#multiple parameter/arguments in lambda function
addition1=lambda x,y,z:x+y+z
addition1(1,2,3)
```

```python
#Map function in lambda : it maps the components of list as an input list
# in a list perform square all the components
```

```python
#Traditional way:
list_item=[1,2,3,4,5,6]
def squareFunction(x):
  sqlist=[]
  for item in x:
   sqlist.append(item**2)
  return sqlist

squareFunction(list_item)
```

```python
#Lambda ways
list_item=[1,2,3,4,5,6]
list(map(lambda x:x**2,list_item)) # map(func, *iterables) --> map object
```
