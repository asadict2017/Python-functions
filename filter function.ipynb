```python
#filter function filter out the items of a list that satisfy the condition
#syntax similar to map function
# filter(function,list)
```

```python
lst=[1,2,3,4,5,6,7,8,9]
def even(num):
  if num%2==0:
    return True
list(filter(even,lst))
```

```python
#apply on lambda function
list(filter(lambda x:x%2==0,lst))
```

```python
#filter in multiple condtions of lambda function
lst=[1,2,3,4,5,6,7,8,9]
list(filter(lambda x:x>=5 and x%2==0,lst))
```

```python
#filter function on dictionary items
person=[
    {'name':'asad','age':39},
    {'name':'john','age':32},
    {'name':'krish','age':40}

]

list(filter(lambda persons:persons['age']>35,person))
```
