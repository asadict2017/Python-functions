```python
#The map() function applies on a given function to all items in an input list and returns a map object.
#This is particularly useful for transforming data in a list comprehensively

#syntax:    map(function,inputlist)
```

```python
def sqFunction(x):
   return x**2

listItem=[1,2,3,4,5,6]
#if want to square the listItem then it may use for loop and append on another list
#map makes it simple

list(map(sqFunction,listItem))
```

```python
#map() function in lambda function
#syntax of lambda function:  lambda arguments:expression
#square of a list

listItem=[1,2,3,4,5,6]
list(map(lambda x:x**2,listItem))
```

```python
#map multiple iterables
#addition among two list components
item1=[1,2,3]
item2=[4,5,6]

list(map(lambda x,y:x+y,item1,item2))
```

```python
#convert list of strings to int
stringlist=['1','2','3','4','5']

list(map(int,stringlist))
```

```python
#do upper case of a given list string
stringList1=["mango","apple","orange"]

list(map(str.upper,stringList1))
```
